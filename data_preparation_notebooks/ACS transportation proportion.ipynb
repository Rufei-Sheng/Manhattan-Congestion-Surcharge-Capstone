```python
import pandas as pd
```

```python
acs = pd.read_csv('acs_commute_data_by_taxi_zone.csv')
acs.head()
```

```python
acs['pt'] = acs['bus'] + acs['subway'] 
acs['subtotal'] = acs['bus'] + acs['subway'] + acs['walked'] + acs['taxi'] 
```

```python
acs.head()
```

```python
sub = acs[['taxi_zone', 'pt', 'walked', 'taxi', 'subtotal']]
sub.head()
```

```python
sub['pt'] = sub['pt'] / sub['subtotal']
sub['walked'] = sub['walked'] / sub['subtotal']
sub['taxi'] = sub['taxi'] / sub['subtotal']
```

```python
sub = sub.dropna()
sub = sub.drop('subtotal', axis=1)
```

```python
sub.taxi_zone = sub.taxi_zone.astype(int)
sub.head()
```

```python
taxi_prop = pd.read_csv('taxi_proportions.csv')
taxi_prop = taxi_prop[['PUlocationID', 'Taxi proportion', 'FHV proportion', 'FHV_shared proportion']]
```

```python
taxi_prop.PUlocationID = taxi_prop.PUlocationID.astype(int)
taxi_prop.head()
```

```python
acs_proportion = pd.merge(sub, taxi_prop, left_on='taxi_zone', right_on='PUlocationID', how='left')
```

```python
acs_proportion.head()
```

```python
acs_proportion['Taxi proportion']  = acs_proportion['Taxi proportion'] * acs_proportion['taxi']
acs_proportion['FHV proportion']  = acs_proportion['FHV proportion'] * acs_proportion['taxi']
acs_proportion['FHV_shared proportion']  = acs_proportion['FHV_shared proportion'] * acs_proportion['taxi']
```

```python
acs_proportion.head()
```
