```python
import numpy as np
import pandas as pd
import geopandas as gpd
import shapely
from fiona.crs import from_epsg
import pysal as ps
import matplotlib.pyplot as plt
import pylab as pl
import seaborn as sns
%pylab inline
```

```python
data1= pd.read_csv('results_original_predict_transportation_choice.csv')
data1.head()
```

```python
data2= pd.read_csv('results_add2.75_predict_transportation_choice.csv')
data2.head()
```

```python
data3= pd.read_csv('results_add10_predict_transportation_choice.csv')
data3.head()
```

```python
data1_sum = data1.groupby(data1.origin).sum()
data1_sum.head()
```

```python
data2_sum = data2.groupby(data1.origin).sum()
data2_sum.head()
```

```python
data3_sum = data3.groupby(data1.origin).sum()
data3_sum.head()
```

```python
difference_between_0_and_2_75 =data2_sum - data1_sum
difference_between_0_and_2_75 = difference_between_0_and_2_75.drop(['destination', 'origin'], axis=1)
difference_between_0_and_2_75['origin'] = difference_between_0_and_2_75.index
difference_between_0_and_2_75.head()
```

```python
differences = pd.DataFrame()
differences['P(mode1)'] = [difference_between_0_and_2_75.abs().sort_values(by ='P(mode1)', ascending=False).iloc[0].origin.astype('int')]
differences['P(mode2)'] = [difference_between_0_and_2_75.abs().sort_values(by ='P(mode2)', ascending=False).iloc[0].origin.astype('int')]
differences['P(mode3)'] = [difference_between_0_and_2_75.abs().sort_values(by ='P(mode3)', ascending=False).iloc[0].origin.astype('int')]
differences['P(mode4)'] = [difference_between_0_and_2_75.abs().sort_values(by ='P(mode4)', ascending=False).iloc[0].origin.astype('int')]
differences['P(mode5)'] = [difference_between_0_and_2_75.abs().sort_values(by ='P(mode5)', ascending=False).iloc[0].origin.astype('int')]
differences
```

```python
difference_between_0_and_10 =data3_sum - data1_sum
difference_between_0_and_10 = difference_between_0_and_10.drop(['destination', 'origin'], axis=1)
difference_between_0_and_10['origin'] = difference_between_0_and_10.index
difference_between_0_and_10.head()
```

```python
differences2 = pd.DataFrame()
differences2['P(mode1)'] = [difference_between_0_and_10.abs().sort_values(by ='P(mode1)', ascending=False).iloc[0].origin.astype('int')]
differences2['P(mode2)'] = [difference_between_0_and_10.abs().sort_values(by ='P(mode2)', ascending=False).iloc[0].origin.astype('int')]
differences2['P(mode3)'] = [difference_between_0_and_10.abs().sort_values(by ='P(mode3)', ascending=False).iloc[0].origin.astype('int')]
differences2['P(mode4)'] = [difference_between_0_and_10.abs().sort_values(by ='P(mode4)', ascending=False).iloc[0].origin.astype('int')]
differences2['P(mode5)'] = [difference_between_0_and_10.abs().sort_values(by ='P(mode5)', ascending=False).iloc[0].origin.astype('int')]
differences2
```

```python
differences_ = pd.concat([differences, differences2], axis=0)
differences_.iloc[0,0] = "7 (Astoria)"
differences_.iloc[1,0] = "7 (Astoria)"
differences_.iloc[0,3] = "7 (Astoria)"

differences_.iloc[0,1] = "79 (East Village)"
differences_.iloc[0,2] = "79 (East Village)"
differences_.iloc[0,4] = "232 (University Heights/Morris Heights)"

differences_.iloc[1,1] = "39 (Canarsie)"
differences_.iloc[1,2] = "228 (Sunset Park West)"
differences_.iloc[1,3] = "203 (Rosedale)"
differences_.iloc[1,4] = "203 (Rosedale)"


differences_['taxi zone with the biggest mode change'] = ['orginal to $2.75','orginal to $10']
differences_=differences_.set_index(['taxi zone with the biggest mode change'])
differences_
```
